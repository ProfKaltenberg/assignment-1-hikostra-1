```python
import requests
import pandas as pd
import numpy as np
```

```python
#Open Meteo API  to make weather data 
```

```python
# i changed my API to this one bacause it has history weather data
#this is open API, so there is no key 
url = "https://archive-api.open-meteo.com/v1/archive?latitude=40.7128&longitude=-74.0060&start_date=2025-01-01&end_date=2026-04-09&daily=temperature_2m_mean,relative_humidity_2m_mean,cloudcover_mean,windspeed_10m_max&timezone=America/New_York"
#the coordinates are NY 
response = requests.get(url)
weather_data = response.json()

print(weather_data.keys())
```

```python
#creating dataframe with daily variables
daily = weather_data["daily"]

weather_df = pd.DataFrame(daily)

weather_df.head()
```

```python
#renaming columns 
weather_df.columns = [
    "date",
    "temperature",
    "humidity",
    "cloud_cover",
    "wind_speed"
]

weather_df.head()
```

```python
weather_df.to_csv("weather_data.csv", index=False)
```

```python
spy_df = pd.read_csv('/Users/liza/Desktop/ECO590/dataspyeco590.csv')
spy_df = spy_df.dropna(subset=["Date"]) 
spy_df["Date"] = pd.to_datetime(spy_df["Date"], format="%d/%m/%y") #check the format of the date 


spy_df.head()
```

```python
#renaming columns 
spy_df.columns = [
    "date",
    "close",
    "change_close",
    "daily_return",
    "open",
    "change_open",
    "return_open",
    "high",
    "change_high",
    "return_high",
    "low",
    "change_low",
    "return_low"
]
spy_df.head()
```

```python
#merging 2 datasets 
spy_df['date'] = pd.to_datetime(spy_df['date'])
weather_df['date'] = pd.to_datetime(weather_df['date'])

merged_df = pd.merge(spy_df, weather_df, on="date", how="inner")

merged_df.head()
```

```python
merged_df = merged_df[["date", "close", "daily_return","temperature","humidity","cloud_cover","wind_speed"]]
merged_df.head()
#renaming columns 
```

```python
#creating new calculaed variable 
```

```python
merged_df["good_weather"] = (
    (merged_df["temperature"].between(5, 22)) &
    (merged_df["cloud_cover"] < 60) &
    (merged_df["wind_speed"] < 25)
).astype(int)
#my criteria for good weather 
merged_df.head()
```

```python
merged_df["good_weather"].value_counts() #to look how many days match my criteria
```

```python
merged_df.to_csv("final_dataset.csv", index=False)
```
